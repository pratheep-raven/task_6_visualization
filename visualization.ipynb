```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Retail data set')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
top10 = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(10)

plt.figure()
plt.bar(top10.index, top10.values)
plt.title("Top 10 Categories by Sales")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales = df.groupby(df['Date'].dt.month)['Total Amount'].sum()

plt.figure()
plt.plot(monthly_sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure()
plt.hist(df['Total Amount'], bins=10)
plt.title("Total Amount Distribution")
plt.xlabel("Total Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Total Amount'], df['Price per Unit']) # Using 'Total Amount' for Sales and 'Price per Unit' as a placeholder for Profit
plt.title("Total Amount vs Price per Unit")
plt.xlabel("Total Amount")
plt.ylabel("Price per Unit")
plt.show()
```
